```python
'''
Q3: Describe how to access, modify and delete elements in a list with examples?
Answer:
    To access the element , modify the element and to delete the element from the list is a straightforward process in python,
    lets understand this all concept with the help of examples.
'''
# Creating a list
fruit_list = ["apple", "pineapple", "peach", "watermelon"]
#This is a list containing fruits
print(fruit_list)
'''
1.Accessing the list
     You can access the element of the list using index number, and index number always starts with 0, Here's how can we do
     
Example:
    For example you want to access whats on the first index simply you can
'''
print(fruit_list[1])
#Output will be pineapple, because pineapple is on index 1

# if you want to access the last element from the list, and you dont know the length of the list,
#you can access through negative indices(index)
print(fruit_list[-1])#In reverse order
#Output will be watermelon,
# because it is the last element of the list and negative index starts from the end of list. Eg (-1)

'''
2.Modifying the list:
    List is mutable data type(structure), this means we can easily change any element from the list, 
    just we have to give index number and assign the new value to the list

Example:
    I will modify the list of fruits, which i made previously
'''
fruit_list[2] = "orange"
# This will replace peach from the list, which is on index 2 and add orange to that position.
print(fruit_list)
# Output will be:['apple', 'pineapple', 'orange', 'watermelon']
# This is how you can modify a list, using index number.

'''
3.Deleting the element from the list
    There are three methods of deleting an element from the list
    
    1st.remove()
        #Purpose: remove() method is to, remove a specific value using their index number
        #Usage: When you know the name of the element, you can use remove specific element with the remove() method
Example:
'''
fruit_list.remove("orange")
print(fruit_list)
# Output will be: ['apple', 'pineapple', 'watermelon'], removing orange from the list
'''
    2nd.pop()
        #Purpose: pop() method also remove or delete the element from the list using index number, but pop() method also returns the value
        #Usage: when you want to remove an element using their index number and you want to store that removed element, you use pop() method to perform that
Example:
'''    
removed_fruit = fruit_list.pop(1)
print(fruit_list)
# Output will be: ['apple', 'watermelon'] , removing pineapple which was on first index in the list
# removed_fruit is the variable, which will store the removed element from the list.

'''
    3rd.del()
        #Purpose: del() method will directly delete the element using index number and it does not return the value.
        #Usage: when you want to delete an element without storing them, you can use del() method to perform that.
Example: 
    same i will use the previous fruits list
'''
del fruit_list[1]
print(fruit_list)
# Output will be: ['apple'] , deleting watermelon from the list, which was on first index.

'''
Insights:
    #Access:If you want to only see the element in the list, you can access the element through index number, data(fruit_list) will be intact.
    #Modify:If you want to change/replace the element in the list, you can directly use index number to update the list removing specific element
    #delete:If you dont want an element in the list, you can delete the element,
        (using) pop() method it will return the value, OR del() method it does not return the value
'''
print()
```
